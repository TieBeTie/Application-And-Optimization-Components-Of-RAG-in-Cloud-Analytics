### Список

**Список** - изменяемый контейнер, позволяющий получить доступ к его элементам по индексу (порядковому номеру). Создается с помощью `[]` (пустой или с элементами) или `list()` (пустой или из другого контейнера):


````python
l1 = []
l2 = list()
l1 == l2
````

Индексация по списку с помощью `[]`:

````python
l3 = [False, 1, '2', [3]]
l3[0], l3[-1]
````

Перезапись первого элемента списка:

````python
l3[0] = True
l3
````

`append(v)` добавляет в конец списка новый элемент:

````python
l3.append('new_element')
l3
````

`remove(v)` удаляет из списка элемент со значением `v` (если `v` несколько - первое слева):

````python
l3.remove('2')
l3
````

### Словарь

**Словарь** - реализация ассоциативного массива (хеш-таблицы) в Python. Создается с помощью `{}` (пустой или с элементами) или `dict()` (пустой или из другого контейнера):

````python
dct1 = {}
dct2 = dict()
dct1 == dct2
````

````python
dct1 = {'Key1' : 'Value1', 'Key2' : 'Value2'}
dct1
````

````python
dct2 = dict([(2, 4), (3, 9)])
dct2
````

````python
dct1['Key3'] = 'Value3'
dct1
````

Удаление элемента словаря:

````python
del dct1['Key1']
dct1
````

`update(dict)` объединяет два словаря в один, результат сохраняется в первый словарь:

````python
dct1.update(dct2)
dct1
````

### Множество

**Множество (set)** - тип и структура данных, которая является реализацией математического объекта множество. Принцип работы также основан на хеш-таблице. Создается с помощью `{}` ( с элементами) или `set()` (пустой или из другого контейнера):

````python
dct1 = {}
set1 = set()
dct1 == set1
````

````python
set1 = {1, 2, 3}
set2 = set([2, 3, 4])
````

`add(v)` добавляет в множество новый элемент (если в множестве уже был этот элемент, то ничего не изменится):

````python
set1.add(5)
set1
````

`discard(v)` удаляет элемент со значением `v` из множества:

````python
set1.discard(2)
set1
````

`update(set)` объединяет два словаря в одно, результат сохраняется в первое множество:

````python
set1.update(set2)
#или
set1 | set2
````

Пересечение двух множеств:

````python
set1 & set2
````

### Цикл for

 Python `for`-цикл устроен следующим образом:

```python
numbers = [1, 2, 3, 5, 7]
for n in numbers:
    print(n)
```

Как можно заметить, в for-выражении отсутствуют привычные элементы. Вместо этого присутствует **итерируемый объект** (`iterable`) - объект, по которому можно итерироваться.

`for`-циклы по спискам, кортежам, строкам:

````python
list_obj = ['list', 1, 2, 3, 4, 5]
for item in list_obj:
    print(item, end=' ')
print()

tuple_obj = ('tuple', 1, 2, 3, 4, 5)
for item in tuple_obj:
    print(item, end=' ')
print()

string_obj = "string12345"
for item in string_obj:
    print(item, end=' ')
print()
````

В случае, если нужно пройтись не по элементам контейнера, а по числам, то можно использовать следующую функцию:

`range([start,] stop[, step])` возвращает неизменяемую последовательность чисел.

По умолчанию `start=0`, `step=1`.

````python
for i in range(len(list_obj)):
    print(i, end = ' ')
````

Также можно воспользоваться функцией `enumerate`, которая создаёт пары индекс и элемент контейнера. По полученному контейнеру можно итерироваться с помощью цикла `for`:

````python
for i, obj in enumerate(list_obj):
    print(f'index: {i} object: {obj}')
````

````python
# распаковка
x, y, z = (1,2,3)
print(x, '_', y, '_', z)
x, y, *z = (1,2,3,4)
print(x, '_', y, '_', z)
x, *y, z = (1,2,3,4)
print(x, '_', y, '_', z)
````

## Генераторы

**Генератор** — это объект, который сразу при создании не вычисляет значения всех своих элементов, а хранит только последний вычисленный элемент, правило перехода к следующему и условие, при котором выполнение прерывается.

Определения:
- `Iterator` - объект с методом `__next__()`
- `Iterable` - объект с методом `__iter__()`, который возвращает `Iterator`

Создавать генераторы можно двумя способами:
1. Генераторные выражения

````python
gen = (x**2 for x in range(5))
type(gen)
````

Обращаться к следующему элементу можно с помощью функции `next()`

````python
next(gen)
````

2. Генераторные функции

В генераторных функциях для возвращения элемента используется ключевое слово `yield`. Оно похоже на `return`, но в отличие от него не завершает работу функции, а лишь приостанавливает её.

````python
def foo(x):
    print('I am the generator!')
    for value in x:
        yield value
````

````python
values = ['Hello', 'world!']

#foo - generator function
#foo() - generator

for value in foo(values):
    print(value)
````

### Чтение файла

Самым удобным и современным способом чтения и записи файлов являются менеджеры контекстов (если в функции `open` не указывать параметры, то файл будет открыт на чтение):

````python
with open('text.txt') as f:
    for s in f:
        print(s[:-1])
````

Получился объект `f` одного из файловых типов. Что с ним можно делать?

Можно его использовать в `for` цикле, каждый раз будет возвращаться очередная строка файла (включая `'\n'` в конце; в конце последней строки текстового файла `'\n'` может и не быть).

### Запись в файл

Таким же образом, но указав режим записи `'w'` в параметрах функции `open` можно произвести запись в файл.
Если файла с таким именем нет, то он создается автоматически:

````python
with open('newtext.txt', 'w') as f:
    f.write('aaa\n')
    f.write('bbb\n')
    f.write('ccc\n')
````

````python
!cat newtext.txt
````

| Режим | Обозначение |
|-------|-------------|
| **'r'**  | Открытие на **чтение** (является значением по умолчанию) |
| **'rb'** | Открытие на **чтение**, в предположении, что будут считываться **байты** |
| **'w'** | Открытие на **запись**, содержимое файла удаляется. Если файла не существует, создается новый |
| **'wb**' | Открытие на **запись байтов**, содержимое файла удаляется. Если файла не существует, создается новый |
| **'a'** | Открытие на **дозапись**, информация добавляется **в конец файла** |
| **'r+'** | Открыть файл на **чтение И запись**. Если файла нет, **новый НЕ создаётся** |
| **'a+'** | Открыть файл на **чтение И запись в конец файла**. Если файла нет, **новый создаётся** |
| **'t'** | Открытие файла **как текстового** (по умолчанию) |

### Объявление класса и экземпляра

В Python объявление класса выполняется следующим образом:

````python
class MyLittleClass(object):  # название класса и родительский класс, родительский класс по умолчанию - object, его можно и не указывать
#class MyLittleClass:
    # Атрибут класса
    color = "blue"

    # Метод
    def set_color(self, new_color):
        color = new_color
        print('set color to', color)
````

Создание объекта - экземпляра класса выглядит так:

````python
obj = MyLittleClass()
type(obj)
````

Обращение к атрибуту экземпляра выглядит следующим образом:

````python
obj.color
````

Вызов метода экземпляра выглядит так:

````python
obj.set_color('red')
obj.color
````

**Почему `color` внутри объекта не поменялся?**

Потому что обращение к атрибутам класса должно иметь форму `self.attribute_name`, а `color` в методе `set_color` -- просто локальная переменная :)

````python
class MyLittleClass2:
    color = "blue"

    def set_color(self, new_color):
        self.color = new_color
        print('set color to', self.color)
````

````python
obj = MyLittleClass2()
obj.color
````

````python
obj.set_color('red')
obj.color
````

### `__str__`



Создадим простой класс: вектор с двумя координатами и цветом.

````python
import random

class Vector:
    def __init__(self, x=0, y=0, color=None):
        print("initializing a vector")
        if type(x) is not int or type(y) is not int:
            raise AttributeError('x and y should be int')

        self._x = x
        self._y = y
        self._color = color

    def get_x(self):
        return self._x

    def get_y(self):
        return self._y
````

Создадим экземпляр вектора и посмотрим на его строчное представление:

````python
vector = Vector(1, 2, 'red')
str(vector)
````

Информативно, но не очень красиво. Есть возможность переопределить поведение метода приведения нашего класса к строковому типу - `__str__`:

````python
class VectorWithStr(Vector):
    def __str__(self):
        return f"vector ({self._x}, {self._y}) of color {self._color}"
````

````python
vector = VectorWithStr(1, 2, 'red')
str(vector)
````

### Арифметические методы (`__abs__`, `__add__`, `__sub__` и другие)

В Python имеются magic-методы для поддержки арифметических операций с пользовательскими классами:

````python
import math
import random

class VectorWithMath(VectorWithRepr):
    def __abs__(self):
        return math.hypot(self._x, self._y)

    def __add__(self, other):
        return VectorWithMath(
            self.get_x() + other.get_x(),
            self.get_y() + other.get_y(),
            random.choice((str(self._color), str(other._color)))
        )

    def __sub__(self, other):
        return VectorWithMath(
            self.get_x() - other.get_x(),
            self.get_y() - other.get_y(),
            random.choice((str(self._color), str(other._color)))
        )

    # ещё есть div, mul и многое другое
````

На вход подаётся число N - вывести матрицу-спираль размера NxN. Пример для 3x3:

1 2 3

8 9 4

7 6 5


````python
N = int(input())
matrix = [[0]*N for _ in range(N)]
num = 1
layer = 0
while num <= N*N:
    for i in range(layer, N - layer):
        matrix[layer][i] = num
        num += 1
    for i in range(layer + 1, N - layer):
        matrix[i][N - layer - 1] = num
        num += 1
    for i in range(N - layer - 2, layer - 1, -1):
        matrix[N - layer - 1][i] = num
        num += 1
    for i in range(N - layer - 2, layer, -1):
        matrix[i][layer] = num
        num += 1
    layer += 1
for row in matrix:
    print(' '.join(map(str, row)))
````

И в первом и во втором случае функция должна вернуть отсортированный список.

````python
def process_lists(lst1, lst2):
    if len(lst1) == len(lst2):
        result = [a * b for a, b in zip(lst1, lst2)]
    else:
        combined = lst1 + lst2
        counts = {}
        for item in combined:
            counts[item] = counts.get(item, 0) + 1
        result = [item for item, count in counts.items() if count >= 2]
    return sorted(result)
lst1 = [1, 2, 3, 4]
lst2 = [2, 3, 4, 5]
print(process_lists(lst1, lst2))
````

Задача 15.

Напишите функцию для нахождения самой длинной подстроки одинаковых подряд идущих букв в заданной строке.

````python
def longest_repeating_substring(s):
    if not s:
        return ''
    max_len = 1
    current_len = 1
    max_char = s[0]
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            current_len += 1
            if current_len > max_len:
                max_len = current_len
                max_char = s[i]
        else:
            current_len = 1
    return max_char * max_len

s = "alsjglkjhiojbifffffjribjwrgkrjvkfvjfkbjdbksjgwkkkkwefwk"
print(longest_repeating_substring(s))

````

````python
import math

class Circle:
    def __init__(self, radius, x=0, y=0):
        # Конструктор, инициализирующий радиус и координаты центра (по умолчанию в начале координат)
        self.radius = radius
        self.center = (x, y)

    def get_radius(self):
        # Метод, возвращающий радиус
        return self.radius

    def get_diameter(self):
        # Метод, возвращающий диаметр
        return self.radius * 2

    def __len__(self):
        # Метод __len__, возвращающий длину окружности (целое число)
        return int(2 * math.pi * self.radius)

    def get_area(self):
        # Метод, возвращающий площадь круга
        return math.pi * self.radius ** 2

    def scale_radius(self, factor):
        # Метод, умножающий радиус на заданное число
        self.radius *= factor

    def __mul__(self, other):
        # Метод __mul__, реализующий оператор умножения
        # Возвращает новую окружность с радиусом, умноженным на заданное число
        return Circle(self.radius * other, *self.center)

    def get_center(self):
        # Метод, возвращающий координаты центра
        return self.center

    def move_center(self, dx, dy):
        # Метод, перемещающий центр на заданный вектор (dx, dy)
        x, y = self.center
        self.center = (x + dx, y + dy)

    def distance_to_origin(self):
        # Метод, возвращающий расстояние от центра окружности до начала координат
        x, y = self.center
        return math.hypot(x, y)

    def __add__(self, other):
        # Метод __add__, реализующий оператор сложения
        # Возвращает новую окружность с центром в середине отрезка между центрами данных окружностей
        # Площадь новой окружности равна сумме площадей данных окружностей
        new_x = (self.center[0] + other.center[0]) / 2
        new_y = (self.center[1] + other.center[1]) / 2
        new_area = self.get_area() + other.get_area()
        new_radius = math.sqrt(new_area / math.pi)
        return Circle(new_radius, new_x, new_y)

    def __str__(self):
        # Метод __str__, возвращающий уравнение окружности в стандартном виде
        x, y = self.center
        return f"(x - {x})^2 + (y - {y})^2 = {self.radius ** 2}"
````
